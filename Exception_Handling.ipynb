```python
class Bank:
  def __init__(self,b):
    self.balance = b

  def Withdraw(self,amount):
    if(amount > self.balance):
      raise Exception("Not Enough Balance!")

    if amount <0:
      raise Exception("Amount must be a positive number!")

    self.balance -=amount


b = Bank(1000)

try:
  b.Withdraw(-11)
except Exception as e:
  print(e)

else:
  print(b.balance)

finally:
  print("TYVM for yrs Visit")

```

# Creating a custom Exception

```python
class MyException(Exception):
  def __init__(self, msg):
    print(msg)

class Bank:
  def __init__(self,b):
    self.balance = b

  def Withdraw(self,amount):
    if(amount > self.balance):
      raise MyException("Not Enough Balance!")

    if amount <0:
      raise MyException("Amount must be a positive number!")

    self.balance -=amount


b = Bank(1000)

try:
  b.Withdraw(111111010111101111011)
except MyException as e:
  pass

else:
  print(b.balance)

```

# Simple code to show why we need our own custom Exception class

```python
class SecurityError(Exception):
  def __init__(self,message):
    print(message)

  def logout(self):
    print("Logout")


class Google:
  def __init__(self,name,email,password,device):
    self.name = name
    self.email = email
    self.password = password
    self.device = device

  def login(self, email, password, device):
    if device != self.device:
      raise SecurityError("Sth wrong is happening on yrs account")

    if email == self.email and password == self.password:
      print("Welcome")

    else:
      print("Login Error")


obj = Google("Ram","ram@gmail.com","1234","android")

try:
  obj.login("ram@gmail.com","1234","android")

except SecurityError as e:
  e.logout()

else:
  print(f"Yrs name is: {obj.name}")

finally:
  print("Database connection closed.")
```
